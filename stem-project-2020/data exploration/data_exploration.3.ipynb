# Comprehensive scatterplot

```
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
from scipy.stats import norm
from sklearn.preprocessing import StandardScaler
from scipy import stats
import warnings
warnings.filterwarnings('ignore')
%matplotlib inline
```

```
df_train = pd.read_csv('/content/summarized data.xlsx - 工作表3.csv')
df_train.head()
#df_train.shape
```

```
sns.set()
cols = ['House_Price_Index','House_Price_Index_2Y','House_Price_Index_1Y',
    'Resident_Population_1Y','GDP_1Y','Mortgage_Rate_1Y','Owners_Equivalent_Rent_1Y','CPI_1Y']
sns.pairplot(df_train[cols])
plt.show();
```

# Correlation matrix

```
#correlation matrix
corrmat = df_train.corr()
f, ax = plt.subplots()
sns.heatmap(corrmat, vmax=1, square=True);
```
